```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Connect to the database

engine = create_engine('mysql+pymysql://root:root@123@localhost/covid_exploration')
```

```python
covid_deaths = pd.read_csv('covid-deaths.csv')
covid_deaths
```

```python
covid_vaccinations = pd.read_csv('covid-vaccinations.csv')
covid_vaccinations
```

```python
# Writing the data frames to sql table
```

```python
covid_deaths.to_sql(con=engine, name='covid_deaths', if_exists='replace', index = False)
```

```python
covid_vaccinations.to_sql(con=engine, name= 'covid_vaccinations', if_exists= 'replace', index = False)
```
